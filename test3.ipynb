```python
import nltk , sys,tweepy,csv,re
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C:/Users/KIIT/Documents/minor/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')
```

```python
data.head(3)
```

```python
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
```

```python
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
data['subjectivity'] = data['reviews.text'].apply(getSubjectivity)
```

```python
data.head(3)
```

```python
data['polarity'] = data['reviews.text'].apply(getPolarity)
```

```python
data.head(3)
```

```python
p = 0
n = 0
nt = 0
def getAnalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
data['review_analysis'] = data['polarity'].apply(getAnalysis)
```
